## python function

````
def common(list_a , list_b):
  common_num = []
  
  for a in list_a:
    for b in list_b:
      if a == b:
        common_num.append(a)
  return common_num
````

````
x = [1,2,3,4]
y = [2,4,5,6]
common(x,y)
````


## list comprehension

## `new_list = [expression for_loop_one_or_more condtions]`




````
numbers = [1, 2, 3, 4]
squares = [n**2 for n in numbers]

print(squares)
````

````
list_a = [1, 2, 3, 4]
list_b = [2, 3, 4, 5]

common_num = [a for a in list_a for b in list_b if a == b]

print(common_num)
````

````
list_a = [1, 2, 3]

square_cube_list = [ [a**2, a**3] for a in list_a]

print(square_cube_list)
````

## zipping

````
fruits = ["Apple","banana","Grapes"]
prices = [100, 40, 90]


for fruit,price in zip(fruits, prices):
  print(f" {fruit.ljust(10)}  {price}")
````

## lambda, map and filter in Python


1.   lambda operator or lambda function is used for creating small, one-time and anonymous function objects in Python.
2.   Map executes function for each element in the sequence and returns a list of the elements modified by the function
3.   filter returns only those element for which the function_object returns true


---






```
lambda arguments : expression
```



````
add = lambda x, y : x + y 
  
print(add(2, 3))
````



---





```
map(function_object, iterable1, iterable2,...)
```



````
def multiply2(x):
  return x * 2
num =  [1, 2, 3, 4]

list(map(multiply2,num)) 
````

````
list(map(lambda x : x*2, [1, 2, 3, 4]))
````

````
dict_a = [{'name': 'python', 'points': 10},
          {'name': 'java', 'points': 8}]
  
out1 =map(lambda x : x['name'], dict_a) # Output: ['python', 'java']
  
out2 = map(lambda x : x['points']*10,  dict_a) # Output: [100, 80]

out3 = map(lambda x : x['name'] == "python", dict_a) # Output: [True, False]

print(list(out1))
print(list(out2))
print(list(out3))
````

````
a = [1, 2, 3, 4, 5, 6]
list(filter(lambda x : x % 2 == 0, a))
````
